```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("marketing_campaign.csv", sep=";")
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df["Income"] = df["Income"].fillna(df["Income"].median())
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(2026 - df["Year_Birth"], bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Income"], bins=30)
plt.title("Income Distribution")
plt.show()
```

```python
plt.Figure(figsize=(8, 5))
sns.countplot(data=df, x="Marital_Status")
plt.xticks(rotation=45)
plt.title("Marital Status Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Education")
plt.xticks(rotation=45)
plt.title("Education Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["MntWines"], bins=20)
plt.title("Wine Purchase")
plt.show()
```

```python
plt.figure(figsize=(12,10))

sns.heatmap(
    df.select_dtypes(include="number").corr(),
    annot=False,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
)

plt.title("Correlation Heatmap", fontsize=18)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Response")
plt.title("Campaign Distribution")
plt.show()
```

```python
df["Total_Purchases"] = (df["MntWines"] + df["MntFruits"] + df["MntMeatProducts"] + df["MntFishProducts"] + df["MntSweetProducts"] + df["MntGoldProds"])
plt.figure(figsize=(8, 5))
sns.histplot(df["Total_Purchases"], bins=20)
```

# Business Insights

## Key Findings

- Customers with higher income tend to spend more across product categories.
- Wine products account for the highest customer expenditure compared to other product categories.
- Most customers did not respond positively to the marketing campaigns, indicating room for improving campaign effectiveness.
- Married customers form one of the largest customer segments in the dataset.
- Customer spending is positively correlated across multiple product categories.
- Missing values in the Income column were handled using the median value.
- Duplicate records were removed to improve data quality before analysis.

## Recommendations

- Focus premium marketing campaigns on high-income customers.
- Increase promotional offers for customers with low purchase amounts.
- Improve campaign targeting using customer demographics and purchasing history.
- Develop personalized offers based on customers' buying patterns.
